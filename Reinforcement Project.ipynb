```python
import pandas as pd
```

```python
dataset = pd.read_csv('housing.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset.fillna({'price':0,'sqft_living':0,'sqft_lot':0,'yr_built':0,'city':'Others'},inplace=True)
dataset
```

```python
dataset.isnull().sum()
```

```python
cleaned_dataset = dataset.rename(columns={'date':'Date','price':'Price','bedrooms':'Bedrooms','bathrooms':'Bathrooms','sqft_living':'Sqft_living','sqft_lot':'Sqft_lot',
                  'floors':'Floors','waterfront':'Waterfront','view':'View','condition':'Condition','sqft_above':'Sqft_above','sqft_basement':'Sqft_basement',
                  'yr_built':'Year_Built','yr_renovated':'Year_renovated','street':'Street','city':'City','statezip':'Statezip','country':'Country'})
cleaned_dataset
```

```python
cleaned_dataset['Date'] = pd.to_datetime(cleaned_dataset['Date'], format='%d/%m/%Y %H:%M').dt.date
cleaned_dataset
```

```python
cleaned_dataset['Year_Built'] = cleaned_dataset['Year_Built'].astype(int)
cleaned_dataset
```

```python
cleaned_dataset['Total_Sqft_Area'] = cleaned_dataset['Sqft_living'] + cleaned_dataset['Sqft_lot']
cleaned_dataset.head(10)
```

```python
cleaned_dataset['Year_Difference'] = cleaned_dataset.apply(lambda row: row['Date'].year - row['Year_Built'] if row['Year_Built'] != 0 else 0, axis=1)
cleaned_dataset
```

```python
cleaned_dataset['Current Age of House'] = cleaned_dataset.apply(lambda row: 2025 - row['Year_Built'] if row['Year_Built'] != 0 else 0, axis=1)
cleaned_dataset
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cleaned_dataset
```

```python
import pandas as pd
import seaborn as sns
# Columns to check for outliers
columns = ['Price', 'Bedrooms', 'Bathrooms', 'Sqft_living', 'Sqft_lot', 
                    'Floors', 'Waterfront', 'View', 'Condition', 'Sqft_above', 
                    'Sqft_basement', 'Year_Built', 'Year_renovated']

for col in columns:
    Q1 = cleaned_dataset[col].quantile(0.25)
    Q3 = cleaned_dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_dataset = cleaned_dataset[(cleaned_dataset[col] >= lower_bound) & (cleaned_dataset[col] <= upper_bound)]
#Univariate Analysis
#Count Plot
sns.boxplot(data=cleaned_dataset['Price'],color = 'red')
plt.title('Box Plot')
plt.show()
```

```python
#Bar plot
top5_cities = cleaned_dataset['City'].value_counts().nlargest(5).index
df_top5 = cleaned_dataset[cleaned_dataset['City'].isin(top5_cities)]
plt.figure(figsize=(8, 5))
sns.barplot(data=df_top5, x='City', y='Bedrooms', palette=['red','yellow','orange','green','blue'])

plt.title('Top 5 Cities')
plt.xlabel('City')
plt.ylabel('Bedrooms')
plt.tight_layout()
plt.show()
```

```python
top5_cities = cleaned_dataset['City'].value_counts().nlargest(5).index

df_top5 = cleaned_dataset[cleaned_dataset['City'].isin(top5_cities)]

pivot = pd.crosstab(df_top5['City'], df_top5['Condition'])

pivot = pivot.loc[top5_cities]

pivot.plot(kind='bar',stacked=False,figsize=(10, 6))

plt.title('Stacked Bar Chart of House Condition in Top 5 Cities')
plt.xlabel('City')
plt.ylabel('Number of Listings')
plt.legend(title='Condition')
plt.tight_layout()
plt.show()
```

```python
#Violin Plot
top5_cities = cleaned_dataset['City'].value_counts().nlargest(5).index

df_top5 = cleaned_dataset[cleaned_dataset['City'].isin(top5_cities)]

plt.figure(figsize=(8, 6))
sns.violinplot(data=df_top5, x='City', y='Sqft_basement', palette=['blue','green','red','yellow','orange'])

plt.title('Distribution of Sqft Basement in Top 5 Cities')
plt.xlabel('City')
plt.ylabel('Sqft Basement')
plt.tight_layout()
plt.show()
```

```python
#Multivariate Analysis - more than two variables
#Pair plot

top5_cities = cleaned_dataset['City'].value_counts().nlargest(5).index
df_top5 = cleaned_dataset[cleaned_dataset['City'].isin(top5_cities)]
sns.pairplot(df_top5, vars=['Sqft_basement', 'Sqft_living', 'Bedrooms'], hue='City', palette='Set2')
plt.suptitle('Pairplot of Top 5 Cities', y=1.02)
plt.show()
```

```python
selected_columns = ['Price', 'Bedrooms', 'Bathrooms', 'Sqft_living']

# Step 2: Compute correlation matrix
corr = cleaned_dataset[selected_columns].corr()

# Step 3: Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Heatmap of  Selected Features')
plt.tight_layout()
plt.show()
```

```python
#Clustered Bar Plot
#For 2 categorical and 1 numerical variables
top5_cities = cleaned_dataset['City'].value_counts().nlargest(5).index

df_top5 = cleaned_dataset[cleaned_dataset['City'].isin(top5_cities)]

plt.figure(figsize=(8, 6))
sns.barplot(x='City',y='Sqft_lot',hue='Country',data=df_top5,palette=['red','blue','green','yellow','black'])
plt.title('Clustered Bar Plot')
plt.show()
```
